## Case study part 1

### 1. Assimilate the business scenario and articulate testable hypotheses

#### Business Scenario: 
AVAIL is changing the business model to an a la carte subscription model. This makes prediction (projection) of revenues harder than before. This model had been tested in various countries for years, thus training data exists. But as now this should be rolled out to the US as well, there is a need to predict the revenue - for planing purposes (staffing and budgeting). Goal is to predict at any time the revenue for the following month - for all countries and on a per country level for the top 10 countries in terms of revenue
#### Hypotheses:
Switching to the new model will increase revenue (and prediction will be more accurate than done by business)

### 2. State the ideal data to address the business opportunity and clarify the rationale for needing specific data

We would need data from countries that already have the a la carte service with attributes about the subscribers that are matchable to those from the US. This would allow us to infere from the other countries to the US. Ideally we see what subsricption they had befor they switched to a la carte - so that we can project the future usage on a per user level. Thus one data set would be users (in a country) and the subscripitons/choices over time together with the revenue they generated.
For the users (subscribers) we need at least those attributes that business assesses as relevant for their purchase behaviour, but in general the more attributes we have, the better. Same for countries. Assuming their could be a difference in the behaviour between countries, we need to have the attributes that could explain such differences (e.g. competitive offers, employment rate, gdp,..)
In addition it would help if we get the historical estimates of business and their accuracy (meaning also the past revenues). Thus a second data set would include the business estimates (over all and per country) together with the real revenues.
In order to assess the business value, it would be necessary to have data regarding the effect of the planning accuracy to the business (staffing, budgeting)

Besdides the basic model RandomForestRegressor we will use standard timeseries models like autoregressive models (AR) and univariate autoregressive moving average models (AR(I)MA). There is not enough data for tensorflow models
Using Prophet could make sense as the characteristics where it applies well are:
* hourly, daily, or weekly observations with at least a few months (preferably a year) of history
* strong multiple “human-scale” seasonalities: day of week and time of year
* important holidays that occur at irregular intervals that are known in advance (e.g. the Super Bowl)
* a reasonable number of missing observations or large outliers
* historical trend changes, for instance due to product launches or logging changes
* trends that are non-linear growth curves, where a trend hits a natural limit or saturates

### 2.Iterate on your suite of possible models by modifying data transformations, pipeline architectures, hyperparameters and other relevant factors.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels.tsa.ar_model import AutoReg, ar_select_order
from statsmodels.tsa.api import acf, pacf, graphics
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.api import qqplot
````

````python
sns.set_style('darkgrid')
pd.plotting.register_matplotlib_converters()
# Default figure size
sns.mpl.rc('figure',figsize=(16, 6))
````

````python
# take data for uk only for testing the models
from cslib import engineer_features
for country, df in ts_data.items():
    if country=="united_kingdom":
        X,y,dates = engineer_features(df)
````

````python
X.plot(figsize=(12,8));
````

````python
# modeL: random forest
from sklearn import svm
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
 ## Perform a train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,shuffle=True, random_state=42)
 ## train a random forest model
param_grid_rf = { 'rf__criterion': ['mse','mae'],'rf__n_estimators': [10,15,20,25]}
pipe_rf = Pipeline(steps=[('scaler', StandardScaler()),
                              ('rf', RandomForestRegressor())])
    
grid = GridSearchCV(pipe_rf, param_grid=param_grid_rf, cv=5, iid=False, n_jobs=-1)
grid.fit(X_train, y_train)
y_pred = grid.predict(X_test)
eval_rmse_rf =  round(np.sqrt(mean_squared_error(y_test,y_pred)))
print("RMSE random forest: ", eval_rmse_rf)
````

````python
#model: auto regression
ag_mod = AutoReg(y,exog=X,lags=0).fit()
y_pred_ag=ag_mod.predict()
eval_rmse_ag =  round(np.sqrt(mean_squared_error(y,y_pred_ag)))
print("RMSE auto regression: ", eval_rmse_ag)
print(ag_mod.summary())
````

````python
#model: arima
arma_mod300 = ARIMA(y,X, order=(3, 0, 0)).fit()
arma_mod300.save('uk_mod')
y_pred_arma=arma_mod300.predict()
eval_rmse_arma =  round(np.sqrt(mean_squared_error(y,y_pred_arma)))
print("RMSE ARIMA: ", eval_rmse_arma)
print(arma_mod300.summary())
````

ARIMA has the best performance here
